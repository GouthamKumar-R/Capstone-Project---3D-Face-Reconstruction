```python
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
```

```python
os.getcwd()
```

```python
fname=[]
count=0
for root, dirs, files in os.walk("C:\\Users\\Thinkpad-E470\\Downloads\\titanic\\Capstone"):
    for file in files:
        if file.endswith(".mat"):
             print(os.path.join(root, file))
             fname.append(os.path.join(root, file))
             count=count+1
```

```python
import random
xf=np.empty([53215,])
yf=np.empty([53215,])
zf=np.empty([53215,])
xcount=0
ycount=0
zcount=0
for i in range(0,count):
    data = scipy.io.loadmat(fname[i])
    X = data['shape'][0]
    Y = data['shape'][1]
    Z = data['shape'][2]
    xrand=random.randint(1, 10)
    yrand=random.randint(1, 10)
    zrand=random.randint(1, 10)
    xcount=xcount+xrand
    ycount=ycount+yrand
    zcount=zcount+zrand
    xf=xf+(X*xrand)
    yf=yf+(Y*yrand)
    zf=zf+(Z*zrand)
```

```python
xf1=xf/xcount
yf1=yf/ycount
zf1=zf/zcount
```
